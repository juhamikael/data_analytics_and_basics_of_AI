Task 2

```python
# number of employees by age group:
def barchart_age(dataframe):
    dataframe["age"] = pd.to_datetime("today").year - pd.to_datetime(dataframe["bdate"]).dt.year
    dataframe["age_group"] = pd.cut(dataframe["age"],
                                    bins=[15, 20, 25, 30, 35, 40, 45, 50, 55, 60, 65, 70],
                                    labels=["20", "25", "30", "35", "40", "45", "50", "55", "60", "65", "70"])
    age_group = dataframe["age_group"]
    age_group_count = age_group.value_counts()
    plt.bar(age_group_count.index, age_group_count.sort_index())
    return dataframe


new_df = barchart_age(emp_dep)
```

```python
def get_percentage(dataframe):
    genders = {"male": 0, "female": 0, "total": 0}
    for index, row in dataframe.iterrows():
        if row["gender"] == 0:
            genders["male"] += 1
        elif row["gender"] == 1:
            genders["female"] += 1
        genders["total"] += 1
    females = round(((genders["female"] / genders["total"]) * 100), 2)
    males = round(((genders["male"] / genders["total"]) * 100), 2)
    values = np.array([males, females])
    labels = ["Males", "Females"]
    plt.pie(x=values, labels=labels, autopct='%1.1f%%', radius=2, explode=(0, 0.05))

percentage = get_percentage(emp_dep)
```

```python
def get_gender_by_age_group(dataframe):
    dataframe_copy = pd.DataFrame(dataframe,columns=["gender","age_group"])
    females = {}
    males = {}

    # Create females and males dict which we use later on
    for index,row in dataframe_copy.iterrows():
        females[(row["age_group"])] = 0
        males[(row["age_group"])] = 0

    for index,row in dataframe_copy.iterrows():
        if row["gender"] == 0:
            males[row["age_group"]]+=1
        else:
            females[row["age_group"]]+=1
    return {"females":females,"males":males}

def show_bar(data_dict:dict):
    females = pd.DataFrame.from_dict(data_dict["females"],orient="index")
    males = pd.DataFrame.from_dict(data_dict["males"],orient="index")
    x_axis = np.arange(len(females))
    plt.bar(x_axis-0.1,males[0],0.2, label = 'Boys')
    plt.bar(x_axis+0.1,females[0],0.2,label = 'Girls')
    plt.ylabel("Lukumäärä")
    plt.xlabel("Ikäryhmä")
    plt.xticks(x_axis, data_dict["females"].keys())
    plt.yticks([n for n in range(3)])
    plt.legend()
    plt.show()

split = get_gender_by_age_group(new_df)
show_bar(split)
```
