```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

emp_dep = pd.read_csv("emp-dep.csv")
emp_dep
```

```python
def scatter_chart_age_salary(dataframe):
    dataframe["age"] = pd.to_datetime("today").year - pd.to_datetime(dataframe["bdate"]).dt.year
    plt.scatter(emp_dep["age"], emp_dep["salary"])


scatter_chart_age_salary(emp_dep)
```

```python
def barchart_salary(dataframe):
    dname_count = dataframe["dname"].value_counts()
    print(dname_count.index)
    plt.bar(height=dname_count.index, x=dname_count)


barchart_salary(emp_dep)
```

```python
# number of employees by age group:
def barchart_age(dataframe):
    dataframe["age"] = pd.to_datetime("today").year - pd.to_datetime(dataframe["bdate"]).dt.year
    dataframe["age_group"] = pd.cut(dataframe["age"],
                                    bins=[15, 20, 25, 30, 35, 40, 45, 50, 55, 60, 65, 70],
                                    labels=["20", "25", "30", "35", "40", "45", "50", "55", "60", "65", "70"])
    age_group = dataframe["age_group"]
    age_group_count = age_group.value_counts()
    plt.bar(age_group_count.index, age_group_count.sort_index())
    return dataframe


barchart_age(emp_dep)
```

```python
def get_percentage(dataframe):
    genders = {"male": 0, "female": 0, "total": 0}
    for index, row in dataframe.iterrows():
        if row["gender"] == 0:
            genders["male"] += 1
        elif row["gender"] == 1:
            genders["female"] += 1
        genders["total"] += 1
    females = round(((genders["female"] / genders["total"]) * 100), 2)
    males = round(((genders["male"] / genders["total"]) * 100), 2)
    values = np.array([males, females])
    labels = ["Males", "Females"]
    plt.pie(x=values, labels=labels, autopct='%1.1f%%', radius=2, explode=(0, 0.05))


percentage = get_percentage(emp_dep)

```
