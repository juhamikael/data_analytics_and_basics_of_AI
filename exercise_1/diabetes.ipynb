```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('diabetes.csv')
df.describe(include="all").round(2)
```

```python
class Histogram:
    """
    Tehtv 2. Piirr hisrogrammi kuvaaja datasta:
    """
    def __init__(self, dataframe):
        self.df = dataframe

    def _histogram(self, column_name):
        plt.figure(figsize=(10, 5))
        plt.hist(self.df[column_name])
        plt.xlabel(column_name)
        plt.ylabel('Frequency')
        plt.show()

    def run_histogram(self):
        columns = []
        for df_column in df.columns:
            columns.append(df_column)
        for column in columns:
            self._histogram(column)


histogram = Histogram(df)
histogram.run_histogram()
```

```python
def correlation_heatmap(dataframe):
    """
    Tehtava 3. Piirr korrelaatioheatmap datasta:
    """
    plt.figure(figsize=(15, 15))
    corr = dataframe.corr()
    sns.heatmap(corr, annot=True)
    plt.show()
correlation_heatmap(df)
```

```python
def potilaiden_lukumaara(dataframe: pd.DataFrame):
    """
    Laske potilaiden lukumr in mukaan siten, ett suurin lukumr on ensin.
    Tulosta mys montako diabetestapausta (1) ja ei diabtesta (0) on aineistossa.
    """
    groupby = dataframe.groupby('Age').size().sort_values(ascending=False)
    plt.rc('font', size=20)
    plt.yticks(np.arange(0, groupby.max() + 1, 5))
    groupby.plot(kind='bar', figsize=(32, 16))
    print(f"Potilaiden lukumr in mukaan\n{'-' * 50}")
    print(groupby)
    print(f"{'-' * 50}")

potilaiden_lukumaara(df)

print("Diabetestapaukset (1) ja ei diabtesta (0) aineistosta")
print(f"{'-' * 50}")
print(df.Outcome.value_counts())
```

```python
def null_values(dataframe):
    """
    Tehtv 5. Onko aineistossa nan arvoja? Jos on, miss sarakkeissa ja montako?
    """
    print("Null Values")
    print(f"{'-' * 50}")
    print(dataframe.isnull().sum())
    print(f"{'-' * 50}")

    print("NaN Values")
    print(f"{'-' * 50}")
    print(dataframe.isna().sum())
null_values(df)
```
